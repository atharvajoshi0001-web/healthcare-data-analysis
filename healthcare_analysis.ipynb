```python
df = pd.read_csv("patient_data.csv")

# Quick look at data
df.head()
```

# Disease-wise patient count

```python

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Disease", palette="Set2")
plt.title("Number of Patients by Disease")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Average treatment cost by disease

```python

avg_cost = df.groupby("Disease")["Treatment_Cost"].mean().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=avg_cost.index, y=avg_cost.values, palette="Set1")
plt.title("Average Treatment Cost by Disease")
plt.ylabel("Cost")
plt.xlabel("Disease")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Hospital stay duration by disease (Box Plot)

```python

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Disease", y="Hospital_Stay_Days", palette="Pastel1")
plt.title("Hospital Stay Duration by Disease")
plt.ylabel("Days")
plt.xlabel("Disease")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Age distribution of patients

```python

plt.figure(figsize=(7,5))
sns.histplot(df['Age'], bins=10, kde=True, color="skyblue")
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.tight_layout()
plt.show()
```

# Gender-wise average treatment cost

```python

gender_cost = df.groupby("Gender")["Treatment_Cost"].mean()
plt.figure(figsize=(5,4))
sns.barplot(x=gender_cost.index, y=gender_cost.values, palette="Set2")
plt.title("Average Treatment Cost by Gender")
plt.ylabel("Cost")
plt.xlabel("Gender")
plt.tight_layout()
plt.show()
```

# Age vs Treatment Cost

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="Age", y="Treatment_Cost", hue="Disease", palette="Set2", s=100)
plt.title("Age vs Treatment Cost by Disease", fontsize=16, fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Treatment Cost")
plt.tight_layout()
plt.show()
```

# Gender Distribution

```python
gender_count = df["Gender"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_count, labels=gender_count.index, autopct="%1.1f%%", startangle=90, colors=["#66b3ff","#ff9999"], shadow=True)
plt.title("Patient Distribution by Gender", fontsize=16, fontweight="bold")
plt.show()
```
